## Visualize

Create any visualizations for your EDA here. Make note in the form of code comments of what your thought process is for your visualizations.

```python
sns.jointplot(x='Profit',y='Marketing',data=sales_df, kind='hex', height=5, ratio=2, marginal_ticks=True)
# Visualization to see how peaks in Marketing also have peaks in Profit.
```

```python
fig, ax = plt.subplots(figsize=(15,6))
g = sns.lineplot(x='Date',  y='Marketing', 
data=sales_df, ax=ax,
             marker='o', markersize=10)
ax2 = ax.twinx()
g = sns.barplot(x='Date',  y='Profit', data=sales_df, ax=ax2,alpha=.5)
ax.set_title('Profit vs Marketing Costs', fontsize = 17)
plt.show()
#Great visualization of profit vs marketing spent!
```

```python
fig, ax = plt.subplots(figsize=(15,6))
g = sns.lineplot(x='State',  y='Marketing', 
data=sales_df, ax=ax,
             marker='o', markersize=10)
ax2 = ax.twinx()
g = sns.barplot(x='State',  y='Profit', data=sales_df, ax=ax2,alpha=.5)
ax.set_title('Profit vs Marketing Costs', fontsize = 17)
plt.show()
# Visualization of state profits vs marketing spent in those states.
```
